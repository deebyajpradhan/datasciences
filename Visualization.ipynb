```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#this allows you see to see the plot created inside the jupyter notebook
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# Functional method to create a matplotlib plot
plt.plot(x,y,'b-')
#Matlab y- b- r- etc are color and line style
```

```python
plt.xlabel('axis')
plt.ylabel('saxis')
plt.title('Drive')
plt.plot(x,y,'o-')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object-oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('axis')
axes.set_ylabel('saxis')
axes.set_title('Drive')
```

```python
fig=plt.figure() #an empty canvas
axes1=fig.add_axes([0.1,.1,0.8,0.8]) #[left,bottom,width,height]
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('BB')
axes2.plot(y,x)
axes2.set_title('YB')
```

```python
fig,axes=plt.subplots(3,3)
#axes.plot(x,y)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(1,2)
```

```python
#plt.subplots(1,2) is doing add_axes automatically based on rows and columns
fig,axes=plt.subplots(1,2) 
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(1,2) 
axes[0].plot(x,y)
axes[0].set_title('Shit0')
axes[1].plot(y,x) #we can do indexing for the axes
axes[1].set_title('Shit1')
```

Figure Size, Aspect ratio and DPI(dots per inch)

```python
fig=plt.figure(figsize=(3,2),dpi=100) #figsize width and height in inches
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(3,2),dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#with legend() we can use label text to clarify what plot is what plot
```

```python
plt.figure(figsize=(10,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='line1')
ax.plot(x,x**3,label='line2',color='#FF3399') #RGB hex color codes
ax.legend(loc=(0.15,0.15)) #location codes,loc=0,1,2,3..10
```

```python
#plot appearance
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#lw=linewidth,alpha for transparency of the line,ls=linestyle
ax.plot(x,y,color='#FF3399',lw='2',alpha=0.9,ls='-.',marker='o',markersize=10,
       markerfacecolor='g',markeredgewidth=2,markeredgecolor='black')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF3399',lw='2')
ax.set_xlim([0,1]) #lower bound and upper bound
ax.set_ylim([0,5])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")
axes[0].set_xscale("log")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x,y,lw=2,color='red')
ax.plot(x**2,y,lw=2,color='black')
```
